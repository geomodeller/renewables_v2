```python
import pandas as pd
```

```python
df = pd.read_csv("Dam's data/Big Cliff Dam hourly data/data.csv")
```

```python
df.info()
```

```python
hours = [str(i) for i in range(1,25)]
```

```python
df = df[df["Hour"].isin(hours)]
df.head(30)
```

```python
df['Date'] + ' ' + df['Hour'].astype(str)
```

```python
df['datetime'] = df['Date'] + ' ' + df['Hour'].astype(str) +':00:00'
```

```python
df.loc[df['Hour'] == str(24) ,'datetime'] =  df['Date'] + ' ' + (df['Hour'].astype(int)-1).astype(str) +':59:59'
```

```python
df.head(30)
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], format='%m/%d/%Y %H:%M:%S')
df = df.set_index('datetime')
df.head(30)
```

```python
df.to_csv("Dam's data/Big Cliff Dam hourly data/data_clean.csv")
```
